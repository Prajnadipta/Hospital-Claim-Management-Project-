```python
patient_df=pd.read_sql_query("SELECT * FROM Patient",connection)
patient_df
```

```python
patient_df["Country"]=patient_df['Country'].str.replace(r"\r","")
patient_df
```

```python
cursor=connection.cursor
```

```python
patient_df["Name"]=patient_df["First_name"]+" "+patient_df["Last_name"]
patient_df
```

```python
patient_df=patient_df.drop('First_name',axis=1)
patient_df=patient_df.drop('Last_name',axis=1)
patient_df
```

```python
patient_df["Address"]=patient_df["Address"]+" "+patient_df["Country"]
patient_df=patient_df.drop('Country',axis=1)
patient_df
```

```python
patient_df["DOB"]=pd.to_datetime(patient_df["DOB"])
patient_df
```

```python
from datetime import date
```

```python
today=date.today()
patient_df["Age"]=pd.DataFrame({'Age':today.year-patient_df["DOB"].dt.year})
patient_df
```

```python
patient_df.to_csv('Patient_Transformed.csv')
```

```python
insurer_df=pd.read_sql_query("SELECT * FROM Insurer",connection)
insurer_df
```

```python
insurer_df['Handling_risk']=insurer_df['Company_name']+insurer_df['Insurance_name']
insurer_df=insurer_df.drop('Company_name',axis=1)
insurer_df=insurer_df.drop('Insurance_name',axis=1)
insurer_df
```

```python
insurer_df["Insurer_Term"]=pd.DataFrame({"Insurer_Term":pd.to_datetime(insurer_df["End_date"]).dt.year-pd.to_datetime(insurer_df["Start_date"]).dt.year})
insurer_df
```

```python
insurer_df['Insurance_amount']=pd.DataFrame(insurer_df['Insurance_base_amount']) + pd.DataFrame(insurer_df['Topup_Amount'])
insurer_df=insurer_df.drop('Insurance_base_amount',axis=1)
insurer_df=insurer_df.drop('Topup_Amount',axis=1)
insurer_df
```
