```python
cursor.execute("select avg(f.Insurance_amount-c.Claim_amount) as Balance_amount, f.Insurance_Details from Final_Table f JOIN Claim c on f.Claim_ID=c.Claim_ID group by Insurance_Details;")
result=cursor.fetchall
balance_amount=[]
insurance_details=[]
for i in cursor:
    balance_amount.append(i[0])
    insurance_details.append(i[1])
    
print("Balance Amount =",balance_amount)
print("Insurance Details =",insurance_details)

plt.rcParams['figure.figsize']=[6,8]
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.bar(insurance_details,balance_amount)
plt.ylim(0, 600000)
plt.xlabel("Insurance Details")
plt.ylabel("Balance Amount")
plt.title("Average Balance Amount per Insurance Company")
plt.show()
```
